```python
import pandas as pd
```

```python
main_data = pd.read_csv("../titanic/train.csv")
```

```python
main_data.head(5)
```

# List of useless columns
- PassengerId
- Ticket
- Name
- Cabin

```python
cleaned_data = main_data.drop(["PassengerId", "Ticket", "Name"], axis=1)
cleaned_data.head(5)
```

```python
cleaned_data = cleaned_data.drop(["Cabin"], axis=1)
```

# Removing Colums with string data using one_hot_encoding
- Sex
- Embarked

```python
# Removing Sex
gender = cleaned_data["Sex"]
male,female = [],[]
for i in gender:
    if i == "male":
        male.append(1)
        female.append(0)
    else:
        male.append(0)
        female.append(1)
```

```python
cleaned_data.insert(2, "Male", male, True)
cleaned_data.insert(2, "Female", female, True)
```

```python
cleaned_data = cleaned_data.drop(["Sex"], axis=1)
```

```python
# Removing Embarked
embarked = cleaned_data["Embarked"]
C, Q, S = [], [], []
for i in embarked:
    if i == "C":
        C.append(1)
        Q.append(0)
        S.append(0)
    elif i == "Q":
        C.append(0)
        Q.append(1)
        S.append(0)
    elif i == "S":
        C.append(0)
        Q.append(0)
        S.append(1)
    else:
        C.append(0)
        Q.append(0)
        S.append(0)
```

```python
cleaned_data.insert(8, "C", C, True)
cleaned_data.insert(8, "Q", Q, True)
cleaned_data.insert(8, "S", S, True)
```

```python
cleaned_data = cleaned_data.drop(["Embarked"], axis=1)
```

```python
cleaned_data.head()
```

```python
final_data = cleaned_data
final_data.to_csv('data_clean.csv')
```
